```python
Name=[]
Model=[]
Buy_Year=[]
Location=[]
Price=[]
Engine_CC=[]
No_of_owners=[]
color=[]
KM=[]
for i in range(1,21):
    url=f"https://www.bikes4sale.in/used/buy/?page={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    for i in soup.find_all('div',class_="txtBox"):
    #print(i)
        link=re.findall('<a href="(.+)" rel',str(i))
    #print(link[0])
        link="https://www.bikes4sale.in"+str(link[0])
    #print(link)
        pages=requests.get(link)
    #print(pages.text)
    
        link_soup=BeautifulSoup(pages.content,'html.parser')
        link_bs=link_soup.find('div',attrs={"id":"secondCol"})
        name=re.findall("<h1>Used (\w+\s?)",str(link_bs))
        
        Name.append(name)
    
        data=link_bs.find_all("table",class_="b4stable itemDtlsTbl")
    
        model=re.findall(": (.+)</td>",str(data))
    
        Model.append(model[0])
        Buy_Year.append(model[1])
    
        Location.append(model[2])
    #print(Location)


        price=link_soup.find('div',class_="specLnk specLnkEmi serLnks")
        pp=re.findall(": Rs. (\d.+)C",price.text)
        if pp:
            Price.append(pp)
        else:
            Price.append(np.nan)
    #print(Price)
    



        link_bss=link_soup.find_all("table",class_="table idNewTbl")
        cc=re.findall("<td>(.+cc)",str(link_bss))
        if cc:
        
            Engine_CC.append(cc[0])
        else:
            Engine_CC.append(np.nan)
    #print(Engine_CC)

        owners = re.findall("<td>(.+) Owner", str(link_bss))
        if len(owners) >= 2:
            No_of_owners.append(owners[1])
        elif len(owners) == 1:
            No_of_owners.append(owners[0])
        else:
            No_of_owners.append(np.nan)

        link_color_table=link_soup.find_all('div',class_="addInfo itemSubDetails")
        
        kms=re.findall("<td>(.+)kms",str(link_color_table))
        if kms:
            KM.append(kms[0])
        else:
            KM.append(np.nan)


df=pd.DataFrame({"Vechicle Name":Name,"Model":Model,"Buy_Year":Buy_Year,"Location":Location,"Price":Price,"Engine_CC":Engine_CC,"No_of_owners":No_of_owners,"KM runs":KM})
df    
    
```

```python
# import pandas as pd
# import re

# # Load your dataset
# df = pd.read_csv("dataset.csv")

# Define a function to extract 3-digit engine cc from model name
def extract_engine_cc(row):
    model = str(row["Model"])
    current_cc = row["Engine_CC"]
    patterns=[r"\b(\d{3})[^a-z]\b",r"\b(\d{3})\b",r"\b\w.+(\d{3})\b"]
    if pd.isna(current_cc):
        for i in patterns:
            match = re.search(i, model)
            if match:
                return f"{match.group(1)} cc"
    return current_cc

# Apply the function row-wise
data["Engine_CC"] = data.apply(extract_engine_cc, axis=1)

# Optional: check how many were filled
print(f"Missing Engine_CC after filling: {data['Engine_CC'].isna().sum()}")
```

```python
Brand=[]
Model=[]
Price=[]
Location=[]
for i in soup.find_all('div',class_="txtBox"):
    txt=i.text
    brand=re.findall("(\w.+), \d{4}",txt)
    if brand:
        Brand.append(brand)
    else:
        Brand.append([np.nan])


    model=re.findall("(\d{4}) ",txt)
    if model:
        Model.append(model)
    else:
        Model.append([np.nan])


    price=re.findall("Rs. (\d.+)",txt)
    if price:
        Price.append(price)
    else:
        Price.append([np.nan])



    location=re.findall(": (\w.+)\s? ",txt)
    if location:
        Location.append(location)
    else:
        Location.append([np.nan])

    
```
